```python
import cv2
import numpy as np
import time
```

```python
cap = cv2.VideoCapture('/Users/mehravehahmadi/yolov5/RPReplay_Final1699467350.MP4')  # Replace with the path to your video

# Load the template image
template_mag = cv2.imread('/Users/mehravehahmadi/yolov5/mag_s.png', cv2.IMREAD_COLOR)  # Replace with the path to your template image
template_sneaker = cv2.imread('/Users/mehravehahmadi/yolov5/snk_s.png', cv2.IMREAD_COLOR)  # Replace with the path to your template image
template_jet = cv2.imread('/Users/mehravehahmadi/yolov5/jet_temp.jpg', cv2.IMREAD_COLOR)  # Replace with the path to your template image
template_2x = cv2.imread('/Users/mehravehahmadi/yolov5/2x_temp.png', cv2.IMREAD_COLOR)  # Replace with the path to your template image
template_sk = cv2.imread('/Users/mehravehahmadi/yolov5/sk_temp.png', cv2.IMREAD_COLOR)  # Replace with the path to your template image

# Get the width and height of the template
template_height_mag, template_width_mag, _ = template_mag.shape
template_height_jet, template_width_jet, _ = template_jet.shape
template_height_sneaker, template_width_sneaker, _ = template_sneaker.shape
template_height_2x, template_width_2x, _ = template_2x.shape
template_height_sk, template_width_sk, _ = template_sk.shape
```

```python
def main():

    mag_time = 0
    jet_time = 0
    sneaker_time = 0
    twox_time = 0
    sk_time = 0
    start_time = time.time()
    while True:
        ret, frame = cap.read()

        if not ret:
            break

        frame1 = frame[1700: 2000 , 200:500]
        frame2 = frame[1700: 2000 , 700:900]

        frame = np.concatenate((frame1, frame2), axis=1)

        #print(max_val_jet)
        #max_result = max(results, key=np.max)
        #max_y, max_x = np.unravel_index(max_result.argmax(), max_result.shape)
       

        #cv2.imshow('Selected ROI', frame)
        #cv2.waitKey(0)
        # Define a threshold to consider a match

        # Perform template matching
        result_meg = cv2.matchTemplate(frame, template_mag, cv2.TM_CCOEFF_NORMED)
        min_val_mag, max_val_mag, min_loc_mag, max_loc_mag = cv2.minMaxLoc(result_meg)

        result_jet = cv2.matchTemplate(frame, template_jet, cv2.TM_CCOEFF_NORMED)
        min_val_jet, max_val_jet, min_loc_jet, max_loc_jet = cv2.minMaxLoc(result_jet)

        result_sneaker = cv2.matchTemplate(frame, template_sneaker, cv2.TM_CCOEFF_NORMED)
        min_val_sneaker, max_val_sneaker, min_loc_sneaker, max_loc_sneaker = cv2.minMaxLoc(result_sneaker)

        result_2x = cv2.matchTemplate(frame, template_2x, cv2.TM_CCOEFF_NORMED)
        min_val_2x, max_val_2x, min_loc_2x, max_loc_2x = cv2.minMaxLoc(result_2x)

        result_sk = cv2.matchTemplate(frame, template_sk, cv2.TM_CCOEFF_NORMED)
        min_val_sk, max_val_sk, min_loc_sk, max_loc_sk = cv2.minMaxLoc(result_sk)

        threshold = 0.6

        if max_val_mag > threshold:
            # Draw a bounding box around the detected region
            top_left = max_loc_mag
            bottom_right = (top_left[0] + template_width_mag, top_left[1] + template_height_mag)
            cv2.rectangle(frame, top_left, bottom_right, (0, 255, 0), 2)
            mag_time += 1
            #print('mag')
        
        if max_val_sneaker > threshold:
            # Draw a bounding box around the detected region
            top_left = max_loc_sneaker
            bottom_right = (top_left[0] + template_width_jet, top_left[1] + template_height_sneaker)
            cv2.rectangle(frame, top_left, bottom_right, (255, 0, 225), 2)
            sneaker_time += 1
            #print('snneaker')

        if max_val_jet > threshold :
            # Draw a bounding box around the detected region
            #top_left = max_loc_jet
            #bottom_right = (top_left[0] + template_width_jet, top_left[1] + template_height_jet)
            #cv2.rectangle(frame, top_left, bottom_right, (0, 0, 225), 2)
            jet_time += 1

        if max_val_2x > threshold:
            # Draw a bounding box around the detected region
            #top_left = max_loc_2x
            #bottom_right = (top_left[0] + template_width_2x, top_left[1] + template_height_2x)
            #cv2.rectangle(frame, top_left, bottom_right, (0, 255, 225), 2)
            twox_time += 1
            #print('2x')

        if max_val_sk > threshold:
            # Draw a bounding box around the detected region
            #top_left = max_loc_sk
            #bottom_right = (top_left[0] + template_width_sk, top_left[1] + template_height_sk)
            #cv2.rectangle(frame, top_left, bottom_right, (0, 255, 0), 4)
            sk_time += 1
            #print('skate')
            

        # Display the frame with the template match
        cv2.imshow('Template Matching', frame)

        if cv2.waitKey(1) & 0xFF == 27:
            break

    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Elapsed time: {elapsed_time} seconds")

    cap.release()
    cv2.destroyAllWindows()
    print('jet = ' , jet_time , 'magnet =' , mag_time , 'sneaker = ' , sneaker_time , '2x = ' , twox_time , 'skate = ' , sk_time)
```

```python
if __name__ == '__main__':
    main()
```
